```python
import matplotlib.pyplot as plt
from matplotlib import rc
import pandas as pd

df = pd.read_csv("./data/netflixRevenue.csv")
df['Netflix Global Users'] = df['Netflix Global Users'].str.replace(',','').astype(float)
df['Year'] = pd.to_datetime(df['Date'], dayfirst = True).dt.year
df['Date'] = pd.to_datetime(df['Date'], dayfirst = True)
df = df.sort_values('Date')
rc('font', family='Apple SD Gothic Neo')

plt.figure(figsize = (13,7))

plt.title("Netflix 전 세계 수익")#, fontdict = kr_ft)
plt.plot(df['Date'], df['Global Revenue']/100000000, '.-')
#plt.xticks(rotation = 45)
plt.grid(True, linestyle = '--')
plt.xlabel("날짜(연)")#, fontdict = kr_ft)
plt.ylabel("수익(단위: 억 달러)")#, fontdict = kr_ft)
plt.savefig('./data/graph/netflixGlobalRevenueYear.png', dpi=200)
plt.show()
```
