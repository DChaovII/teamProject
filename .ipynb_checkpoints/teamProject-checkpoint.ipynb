```python
import matplotlib.pyplot as plt
from matplotlib import rc
import pandas as pd

df = pd.read_csv("./data/netflixRevenue.csv")
df['Netflix Global Users'] = df['Netflix Global Users'].str.replace(',','').astype(float)
df['Date'] = pd.to_datetime(df['Date'], dayfirst = True)
df = df.sort_values('Date')
rc('font', family='Apple SD Gothic Neo')

plt.figure(figsize = (13,7))

plt.title("Netflix 전 세계 수익")#, fontdict = kr_ft)
plt.plot(df['Date'], df['Global Revenue']/100000000, '.-')
#plt.xticks(rotation = 45)
plt.grid(True, linestyle = '--')
plt.xlabel("날짜(연)")#, fontdict = kr_ft)
plt.ylabel("수익(단위: 억 달러)")#, fontdict = kr_ft)
#plt.savefig('./data/graph/netflixGlobalRevenueYear.png', dpi=200)
plt.show()
```

```python
plt.figure(figsize = (13,7))
plt.title('Netflix 전 세계 사용자 수')
plt.plot(df['Date'], df['Netflix Global Users']/100000000, '.-')
plt.grid(True, linestyle = '--')
plt.xlabel('날짜(연)')
plt.ylabel('이용자 수(단위: 억 명)')
#plt.savefig('./data/graph/netfilxGlobalUsers.png', dpi = 200)
plt.show()
```

```python
netWww = pd.read_csv("./data/netflixWhatWeWatched.csv")
netWww = netWww.loc[::-1].reset_index(drop=True)
netWww
```

```python
plt.figure(figsize = (13, 7))

plt.plot(netWww.index[::-1], netWww['Hours Viewed']/10000)
plt.xlabel('순위')
plt.ylabel('시청 시간(단위: 만 시간)')
plt.title('Netflix 순위 별 시청 시간')
plt.grid(linestyle = '--')
#plt.savefig('./data/graph/netflixWhatWeWatched.png', dpi = 200)
plt.show()
```
